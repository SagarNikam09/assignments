Flask is a microframework for Python based on Werkzeug, Jinja2 and good intentions. It is considered to be a microframework because it does not require any particular set of tools or libraries. This makes it very flexible and adaptable to different needs.

Advantages of Flask Framework:

* **Simple and lightweight:** Flask is a very simple framework, which makes it easy to learn and use. It is also very lightweight, which means that it does not require a lot of resources to run.
* **Flexible:** Flask is very flexible, which means that it can be used to build a wide variety of web applications. It is not tied to any particular database or template engine, which gives you a lot of freedom in how you build your application.
* **Extensible:** Flask is extensible, which means that you can add new features and functionality to it as needed. There are a number of extensions available for Flask, which can be used to add features such as user authentication, database integration, and form validation.
* **Community support:** Flask has a large and active community, which means that there is a lot of documentation and support available. This can be helpful if you run into problems or need help with your application.

```python
#Q2]

from flask import Flask 

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Hello, World!</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

#Q3]

App routing in Flask is the process of mapping URLs to functions. This allows you to create different pages or endpoints in your application that can be accessed by different URLs.

App routes are used to control how users interact with your application. They can be used to:

* **Direct users to different pages:** You can use app routes to direct users to different pages in your application, such as a home page, a contact page, or a blog page.
* **Handle different HTTP requests:** You can use app routes to handle different HTTP requests, such as GET requests, POST requests, and PUT requests.
* **Authenticate users:** You can use app routes to authenticate users, so that only authorized users can access certain pages or endpoints.
* **Restrict access to certain pages:** You can use app routes to restrict access to certain pages, so that only users with a certain role or permission can access them.


```python
#Q4]

from flask import Flask 

app = Flask(__name__)

@app.route("/Welcome")
def Welcome():
    return "<h1>Welcome to ABC Corporation</h1>"

@app.route("/")
def Detials():
    return "<h4>Company Name: ABC Corporation <br>Location: India<br>Contact Detail: 999-999-9999</h4>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
