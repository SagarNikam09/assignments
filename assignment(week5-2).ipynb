When we create a custom exception class, we need to inherit from the Exception class. This is because we want our custom exception class to have all of the functionality of the Exception class. For example, our custom exception class will be able to have a message attribute, which is used to store the message of the exception.

```python
#Q2]

import inspect

def print_exception_hierarchy(exception_class):
    print(exception_class.__name__)
    for subclass in exception_class.__subclasses__():
        print_exception_hierarchy(subclass)

print_exception_hierarchy(Exception)
```

The following errors are defined in the ArithmeticError class:

1. ZeroDivisionError: This error is raised when an attempt is made to divide a number by zero.
2. OverflowError: This error is raised when a calculation results in a number that is too large to be represented by the data type.
3. FloatingPointError: This error is raised when a calculation results in a number that is not a valid floating-point number.

```python
def divide_numbers(x, y):
  try:
    return x / y
  except ZeroDivisionError:
    print("Division by zero")

print(divide_numbers(1, 0))
```

#Q4]

The LookupError class is used to handle errors that occur when a value is not found in a sequence or mapping. The KeyError and IndexError exceptions are subclasses of the LookupError class.

```python
dictionary = {"name": "Sagar", "age": 21}

try:
  print(dictionary["address"])
except KeyError:
  print("The key 'address' is not found in the dictionary")
```
